EXPLORATORY DATA ANALYSIS

```python
# To get the count plot of a specific column
graph = sns.countplot(x = 'Size', data = df)
# To provide labels to each bar
for bar in graph.containers:
    graph.bar_label(bar)
    
```

Note : From the above graph most people buy M-Size

```python
# Group by illustration
df.groupby('Size', as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False)
```

```python
plt.figure(figsize=(15,5))
graph2 = sns.countplot(x='Courier Status', hue='Status', data=df)
plt.show()
```

Note : Major of the orders are shipped through couriers

```python
# histogram
df['Size'].hist()
```

```python
df['Size'].hist(bins=15)
```

```python
df['Category'] = df['Category'].astype(str)
col_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(col_data,bins=10, edgecolor='black')
plt.show()
```

Note : From the above graph it is noted that the T-Shirts are maximum purchased

```python
# Checking for B2B buyers and retailers
B2B_check = df['B2B'].value_counts()
plt.pie(B2B_check, labels=B2B_check.index, autopct='%1.1f%%')
plt.show()
```

Note : From the above graph, it is to be noted that the major (99.2%) of the buyers are retailers and only few(0.8%) are B2B buyers.

```python
# Category versus size
x_data = df['Category']
y_data = df['Size']

plt.scatter(x=x_data, y=y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Category VS Sales')
plt.show()
```

```python
# State Distribution
plt.figure(figsize=(12,6))
sns.countplot(x='ship-state', data=df)
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Distribution of States')
plt.xticks(rotation=90)
plt.show()
```

```python
# First 10 States Distribution
plt.figure(figsize=(12,6))
top_states = df['ship-state'].value_counts().head(10)
sns.countplot(data=df[df['ship-state'].isin(top_states.index)], x='ship-state')
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Distribution of States')
plt.xticks(rotation=45)
plt.show()
```

CONCLUSION : 

The data analysis reveals that the business has a significant customer base in Maharashta state, mainly serves retailers, fulfills orders through couriers, experiences high demand for T-Shirts and sees M-Size as the frequent one among buyers.
